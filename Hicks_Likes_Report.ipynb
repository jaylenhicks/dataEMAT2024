**Who or what is the source of your personal Instagram data?**

The source of personal Instagram data is the user. By liking the posts and inteacting with data, users are fundamental in the data curation step. Instagram collects this data and stores it. This data can be used for hypothesis creation and further data analysis.

**Why was this data created and what purposes can the data serve?**

This data was created because it can be extremely beneficial to collect and analyze user data. Data that has been collected over time can be used to show trends. Analyzed trend information can be used to make hypotheses about someone and can push similar content or potential products that the user may be intersted in. 

**In what ways may this data be reliable?**

I think liked data can show reliable trends. From my experiences in how people use instagram, people choose to like posts to show support with their friends and family. Often, when friends and family share a post on Instagram, users will like that post to show the sender that they're engaging with the posts that they send. Users often try to comply with "THE ALGORITHM". Knowing that if they choose to like a certain post, they will be more likely to get similar content to the post that they like. Users spend lots of time interacting with tons of content to try to curate a feed dedicated to the content in which they like. 

**In what ways may this data be unreliable?**

I beleve this data may be unreliable in the conclusions that someone might draw while analyzing the data. Likes are public, so people can see what posts you liked. Knowing this, users may be selective about which posts they choose to like. Information relating to saved posts isn't shared with friends and followers, so I think that saved posts may better reflect which posts users would like to go back and revisit. There are additional features with saved posts, such as the opportunity to create organized collections that may want users to use the save funtion rather than the like function. Since liked posts are stored only in one collection, it may be harder to revisit liked posts.
